```python
def print_name(name, obj):
    if isinstance(obj, h5py.Dataset):
        print('Dataset:', name)
    elif isinstance(obj, h5py.Group):
        print('Group:', name)
```

```python
with h5py.File( 'GEDI02_B_2019137101716_O02416_03_T03984_02_003_01_V002.h5', 'r')  as h5f: # file will be closed when we exit from WITH scope
    h5f.visititems(print_name) # print all strustures names
```
